```python
def link_scrapper():
    """
    Automates web scraping by navigating to a specified URL, scrolling through the page, and saving the HTML content.

    This function uses Selenium WebDriver to open a Chrome browser, navigate to a specified URL (to be replaced by the actual URL),
    scrolls down the page a specified number of times to load dynamic content, then captures the page source HTML. The HTML content 
    is saved to a file and also returned as a string.

    Note: 
    - Requires Selenium WebDriver and ChromeDriver. Ensure ChromeDriver is in the system's PATH or specify its path explicitly.
    - The 'webdriver.ChromeOptions()' is used to disable web security features that might block the script from controlling the browser.
    - The scrolling action is achieved through 'execute_script' with "window.scrollTo" in a loop to load any dynamically loaded content.
    - The final page source is saved to a file 'page_source.txt' and also returned as a string.
    """

    url = 'replace_source_link'
    driver = webdriver.ChromeOptions()
    driver.add_argument("--disable-blink-features=AutomationControlled")
    driver = webdriver.Chrome(os.path.abspath('chromedriver-mac-x64/chromedriver'))
    driver.get(url)
    ScrollNumber = 5
    for _ in range(1, ScrollNumber):
        driver.execute_script("window.scrollTo(1,50000)")

    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')
    result_html = soup.prettify()
    with open('page_source.txt', 'w') as f:
        f.write(result_html)

    return html

# link_scrapper()
```

#### Shuffled Variants.

```python
def shuffle_variants():
    """
    Shuffles tasks within each task file to generate randomized variants for both teachers and students.

    This function iterates over task names derived from a distribution dictionary ('years_distr'), reads tasks from respective files, 
    shuffles them randomly, and then writes two versions of the shuffled tasks into new files: one for teachers including variant 
    numbers and one for students without variant numbers. This is useful for creating randomized task sets to prevent memorization 
    or cheating.

    Note: 
    - Requires the 'random' module for shuffling tasks.
    - Assumes that tasks are stored in text files named after the task names, with tasks separated by 'WORD_FOR_SPLITTING'.
    - The 're' module is used for regex operations to replace placeholder text with specific task numbers or to remove variant numbers 
    for the student version.
    - The path 'path/to/files/' should be replaced with the actual path where task files are stored.
    - Two output files are generated for each task name: '{name}_shuffled.txt' for teachers with detailed task descriptions and 
    variant numbers, and '{name}_shuffled_student.txt' for students with task numbers but without variant numbers.
    """

    names = list(list(years_distr.values())[0].values())
    for name in names:
        file_path = f'path/to/files/{name}.txt'
        with open(file_path) as f:
            all_tasks = ''.join(f.readlines())
            all_tasks = all_tasks.split('WORD_FOR_SPLITTING')
            random.shuffle(all_tasks)
            # Teacher version with Variant Nummer
            with open(f'path/to/files/{name}_shuffled.txt', 'w') as shuffled:
                teacher_tasks = [re.sub('Задание NUMMER', f'Задание {idx + 1}', task) + 'REPLACE' for idx, task in enumerate(all_tasks)]
                shuffled.write('\n\n'.join(teacher_tasks))
            # Student version
            with open(f'path/to/files/{name}_shuffled_student.txt', 'w') as shuffled:
                student_tasks = [re.sub(re.compile('Вариант: \d+<br>replace_link Задание NUMMER'), f'Задание {idx + 1}', task) + 'REPLACE' for idx, task in enumerate(all_tasks)]
                shuffled.write('\n\n'.join(student_tasks))

if int(input('Enter 3')) == 3:
    shuffle_variants()
```
